#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(final_avg_mark):
    if final_avg_mark > 90:
        return 'A'
    elif 80 <= final_avg_mark <= 90:
        return 'B'
    elif 70 <= final_avg_mark < 80:
        return 'C'
    elif 60 <= final_avg_mark < 70:
        return 'D'
    elif 50 <= final_avg_mark < 60:
        return 'E'
    else:
        return 'F'
final_avg_mark = (df["math score"] + df["reading score"] + df["writing score"])/3
df['Grade'] = final_avg_mark.apply(GetGrade)
print("Кол-во учеников получивших A:",len(df[(df["Grade"] == "A")]))
print("Кол-во учеников получивших B:",len(df[(df["Grade"] == "B")]))
print("Кол-во учеников получивших C:",len(df[(df["Grade"] == "C")]))
print("Кол-во учеников получивших D:",len(df[(df["Grade"] == "D")]))
print("Кол-во учеников получивших E:",len(df[(df["Grade"] == "E")]))
print("Кол-во учеников получивших F:",len(df[(df["Grade"] == "F")]))
df
```
